```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#remove outliers
quart1 = data["price"].quantile(0.25)
quart3 = data["price"].quantile(0.75)
IQR = quart3-quart1
l = quart1 - 1.5*IQR
u = quart3 + 1.5*IQR
data_cleaned = data[(data["price"] >= l) & (data["price"] <= u)].copy()
data_cleaned.describe()
```

```python
data_cleaned.info()
```

```python
#change date to usable format
data_cleaned["date"] = pd.to_datetime(data["date"], format="ISO8601")
data_cleaned["year_constructed"] = pd.to_datetime(data["yr_built"],format="%Y")
```

```python
data_cleaned["is_renovated"] = ""
len(data_cleaned["is_renovated"])
```

```python
data_cleaned["is_renovated"] = data_cleaned["yr_renovated"].apply((lambda r: int(bool(r))))
data_cleaned["is_renovated"].head()
```

```python
data_cleaned.iloc[0:15,]
```

```python
#feature engineering - adding years since construction and years since renovation
data_cleaned["years_since_construction"] = data_cleaned["date"].dt.year - data_cleaned["year_constructed"].dt.year
data_cleaned["years_since_construction"]
```

```python
#check for skew in quantitiative variables
skewed_vars = []
for col in data_cleaned.columns:
    if col not in {"date","year_constructed"}:
        print(col,":",data_cleaned[col].skew())
        if (data_cleaned[col].skew())**2 > 1:
            skewed_vars.append(col)
```

```python
#plotting skewed variables dsitribution
fig,axs = plt.subplots(5,2,figsize = (10,10))
for i in range(len(skewed_vars)):
    axs[i//2,i%2].set_title(str("histogram for " + skewed_vars[i]))
    if len(data_cleaned[skewed_vars[i]].unique()) > 100:
        axs[i//2,i%2].hist(data_cleaned[skewed_vars[i]],bins = 50)
    else:
        axs[i//2,i%2].hist(data_cleaned[skewed_vars[i]],bins = 30)

plt.tight_layout()
```

```python
s = np.log(data_cleaned["sqft_above"])
print(s.skew())
```

```python
#from above only need to fix skew for sqft_above since other predictors can be treated as categorical
data_cleaned["log_sqft_above"] = np.log(data_cleaned["sqft_above"])
```

```python
print(data_cleaned.columns)
```

```python
# remove redundant columns
data_processed = data_cleaned.drop(labels=["id","date","year_constructed","sqft_above","yr_built","yr_renovated","lat","long"],axis = 1)
data_processed.columns
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
enc = OneHotEncoder(sparse_output=False)
```

```python
enc.fit(pd.DataFrame(data_processed["zipcode"]))
```

```python
enc.categories_
```

```python
enc.set_output(transform="pandas")
```

```python
zipcode_one_hot = enc.transform(pd.DataFrame(data_processed["zipcode"]))
zipcode_one_hot.head()
```

```python
binary_vars = data_processed.loc[:,["waterfront","is_renovated"]] #save binary variables
data_processed.drop(["waterfront","is_renovated","zipcode"],axis=1,inplace=True) 
```

```python
scaler = StandardScaler() #standardizing data
scaler.set_output(transform="pandas")
scaler.fit(data_processed)
data_standardized = scaler.transform(data_processed)
```

```python
print(data_standardized.shape)
print(zipcode_one_hot.shape)
```

```python
X_y = data_standardized.join([binary_vars,zipcode_one_hot])
print(X_y.shape)
```

```python
y_data = X_y["price"]
X_data = X_y.drop("price",axis = 1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.3, random_state=1)
```

```python
#baseline model
y_bar = np.mean(y_train)
print("training error:",mean_squared_error(y_train,np.array([y_bar]*len(y_train))))
print("baseline testing error:",mean_squared_error(y_test,np.array([y_bar]*len(y_test))))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
```

```python
#model 1 - Ridge regression
ridge = Ridge() 
cv = GridSearchCV(ridge, param_grid={"alpha":[0.1,0.5,1,5,10]},scoring="neg_mean_squared_error")
cv.fit(X_train,y_train)
print(cv.cv_results_)
```

```python
print("ridge regression testing error:",-1*cv.score(X_test,y_test)) 
```

```python
import pymc_bart as pmb
import pymc as pm
```

```python
seed = 123
np.random.seed(seed)
```

```python
X_y.columns
```

```python
#creating BART model
with pm.Model() as mod: #standard syntax for modeling in PyMC
    X = pm.MutableData("X",X_train) #saving as mutable for easy change from training to testing data
    Y = y_train
    mu = pmb.BART("mu",X=X,Y=Y,m=200) #200 tree bart model
    sigma = pm.HalfNormal("sigma", sigma=1) #assuming standard deviation of predictions is half normal
    y_pred = pm.Normal("y_pred",mu=mu,sigma=sigma,observed=y_train,shape = mu.shape) 
    bart_idata = pm.sample(random_seed=seed,tune=100,draws=1000) #1000 burn-in iterations and 1000 draws
    posterior_pred_train = pm.sample_posterior_predictive(trace=bart_idata,random_seed=seed) #sample from posterior
```

```python
posterior_pred_train
```

```python
with mod:
    X.set_value(X_test) #set test data
    posterior_pred_test = pm.sample_posterior_predictive(trace=bart_idata, random_seed=seed) #get sample predictions
```

```python
posterior_pred_test
```

```python
y_test.shape
```

```python
posterior_pred_test.posterior_predictive
```

```python
stacked = posterior_pred_test.posterior_predictive.mean(dim=["chain","draw"]) #take mean over chains and draws
y_pred_test = stacked.y_pred.values
```

```python
y_pred_test.shape
```

```python
print("BART ensemble testing error:",mean_squared_error(y_test,y_pred_test))
```

```python
fig, axs = plt.subplots(1, 2, sharey=True, figsize=(12, 6))

# Ridge Regression
axs[0].scatter(cv.predict(X_test), y_test, alpha=0.5)
axs[0].set_xlabel("Predicted Values")
axs[0].set_ylabel("Actual Values")
axs[0].set_title("Ridge Regression")

# Bayesian Additive Regression Tree
axs[1].scatter(y_pred_test, y_test, alpha=0.5)
axs[1].set_xlabel("Predicted Values")
axs[1].set_ylabel("Actual Values")
axs[1].set_title("Bayesian Additive Regression Tree")

fig.suptitle("Predicted vs Actual Prices on Testing Data")

plt.tight_layout()
plt.show()
```
